# EDA - Exploratory Data Analysis

## Housing Data set

```python
%config InlineBackend.figure_formats = {'png', 'retina'}
```

```python
import pandas as pd
df = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data', header=None, sep='\s+')
```

```python
df.columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
df.head()
```

## Pair Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid', context='notebook')
cols = ['LSTAT', 'INDUS', 'NOX', 'RM', 'MEDV']
sns.pairplot(df[cols], size=2.5, kind='reg')
plt.show()
```

## Joint plot - Histogram & Scatter plot

```python
sns.jointplot('RM', 'MEDV', data=df, kind="reg")
plt.xlabel("RM: Number of rooms (-)")
plt.ylabel("MEDV: Median of husing price (US$)")
plt.show()
```

```python
sns.jointplot('RM', 'MEDV', data=df, kind="kde")
plt.xlabel("RM: Number of rooms (-)")
plt.ylabel("MEDV: Median of husing price (US$)")
plt.show()
```

```python
sns.jointplot('RM', 'MEDV', data=df, kind="hex")
plt.xlabel("RM: Number of rooms (-)")
plt.ylabel("MEDV: Median of husing price (US$)")
plt.show()
```

## Heat Map, Pearson product-moment correlation coefficient

```python
import numpy as np
cm = np.corrcoef(df[cols].values.T)
sns.set(font_scale=1.5)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size':15}, yticklabels=cols, xticklabels=cols)
plt.show()
```

## Iris Data set

```python
iris = sns.load_dataset("iris")
```

```python
pd.set_option("display.max_rows", 10)
iris
```

```python
sns.pairplot(iris, size=2.5, kind='reg', hue="species")
plt.show()
```

```python
cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
cm = np.corrcoef(iris[cols].values.T)
sns.set(font_scale=1.5)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size':15}, yticklabels=cols, xticklabels=cols)
plt.show()
```

## Box plot

```python
sns.boxplot(x='species', y="petal_width", data=iris)
plt.xlabel("Species")
plt.ylabel("Width of Petal (-)")
plt.show()
```

## Bar plot

```python
sns.barplot(x='species', y="petal_width", data=iris)
plt.xlabel("Species")
plt.ylabel("Width of Petal (-)")
plt.show()
```
